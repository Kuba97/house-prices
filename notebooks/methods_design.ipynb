## Grouping neighborhood subject to SalePrice

```python
group_neigh = train[['Neighborhood', 'SalePrice']].groupby(by='Neighborhood')
groupped_neigh_price = group_neigh.describe()['SalePrice']
```

```python
groupped_neigh_price.sort_values(by='mean')
```

```python
intervals = pd.qcut(groupped_neigh_price['mean'], q=8).cat.categories.values
intervals
```

```python
map_dict = {}
new_categories = list('abcdefghijkl')
for index, n in groupped_neigh_price.iterrows():
    found_interval = True
    for i, interv in enumerate(intervals):
        if n['mean'] in interv:
            map_dict[index] = new_categories[i]
            print(index, '->', i)
print(map_dict)
```
