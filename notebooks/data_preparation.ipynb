```python
import os
import pandas as pd
import numpy as np
```

```python
FOLDER_DIR = os.path.join(os.getcwd(), '..', 'data')
TRAIN_DIR = os.path.join(FOLDER_DIR, 'train.csv')
```

```python
def get_raw_data():
    return pd.read_csv(TRAIN_DIR)
```

```python
raw_train = get_raw_data()
```

```python
name_1 ='SalePrice'
name_2 = 'Id'
```

```python
raw_train[raw_train.columns.difference([name_1, name_2])]
```

## NaN's cleaning

```python
cols_to_remove = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'LotFrontage',
        'GarageCond', 'GarageType', 'GarageYrBlt', 'GarageFinish', 'GarageQual',
        'BsmtExposure', 'BsmtFinType2', 'BsmtFinType1', 'BsmtCond', 'BsmtQual',
        'MasVnrArea', 'MasVnrType']
rows_to_remove = raw_train.loc[raw_train['Electrical'].isnull()].index
nan_cleaned = raw_train.drop(cols_to_remove, axis=1)
nan_cleaned = nan_cleaned.drop(rows_to_remove)
# check for null's
print(nan_cleaned.isnull().sum().max()==0)
```

## Log transformation

```python
cols_to_log =  ['SalePrice', 'GrLivArea', '1stFlrSF']
log_transformed = nan_cleaned
log_transformed[cols_to_log] = log_transformed[cols_to_log].apply(np.log)
```

## Dummies

```python
dummied = pd.get_dummies(log_transformed)
```

```python
dataset = dummied
```

## Save dataset

```python
dummied.to_csv(os.path.join(FOLDER_DIR, 'train_prepared.csv'))
```
